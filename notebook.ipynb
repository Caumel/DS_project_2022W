```python
list_names = ["Suhlen1","Suhlen2","Suhlen3","Suhlen4","Suhlen5","Suhlen6","Suhlen7","Suhlen8","Suhlen9","SuhlenFG","SuhlenVG"]
lists = [files_suhlenlogger_1,files_suhlenlogger_2,files_suhlenlogger_3,files_suhlenlogger_4,files_suhlenlogger_5,files_suhlenlogger_6,files_suhlenlogger_7,files_suhlenlogger_8,files_suhlenlogger_9,files_suhlenlogger_FG,files_suhlenlogger_VG]
for index,name in enumerate(list_names):
    result = pd.DataFrame()
    list_data = []
    for file in lists[index]:
        try:
            df = pd.read_csv(os.path.join(path_Suhlenlogger,file), skiprows=13, sep=",",encoding= 'unicode_escape')
            cont = 0
            list = []
            with open(os.path.join(path_Suhlenlogger,file)) as file_obj:
                
                reader_obj = csv.reader(file_obj)
            
                for row in reader_obj:
                    list.append(row[0])
                    if cont == 12:
                        break
                    cont += 1
            
            df["Date/Time"] = pd.to_datetime(df["Date/Time"])
            
            result = pd.concat([result,df.iloc[:,:]])
            list_data.append(list)
        except:
            df = pd.read_csv(os.path.join(path_Suhlenlogger,file), skiprows=18, sep=",",encoding= 'unicode_escape')
            cont = 0
            list = []
            with open(os.path.join(path_Suhlenlogger,file)) as file_obj:
                
                reader_obj = csv.reader(file_obj)
            
                for row in reader_obj:
                    list.append(row[0])
                    if cont == 17:
                        break
                    cont += 1
            
            df["Date/Time"] = pd.to_datetime(df["Date/Time"])
            
            result = pd.concat([result,df.iloc[:,:]])
            list_data.append(list)
    
    with open(os.path.join(path_Suhlenlogger,name + "_summary.csv"), 'w') as f:
        write = csv.writer(f)
        write.writerows(list_data)
    
    result = result.reset_index(drop=True)
    result.to_csv(os.path.join(path_Suhlenlogger, name + ".csv"),index=False)
```
