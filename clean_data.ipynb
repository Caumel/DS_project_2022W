##### Join position with weather

```python
df_env = pd.read_csv(path_env)
df_env = df_env.drop(columns=["Unnamed: 0.1","Unnamed: 0"])
df_env["Date"] = pd.to_datetime(df_env["Date"], infer_datetime_format=True)   
df_env.sort_values(by='Date', inplace=True)
```

```python
path_positions = "../Join/telemetriedaten/positions"
file_positions = os.listdir(path_positions)
for file in file_positions:
    df = pd.read_csv(os.path.join(path_positions,file))
    df = df.drop(columns=["Unnamed: 0"])
    df.columns = ["Date", 'tagid', 'MACadresse', 'x-value',
       'y-value', 'ID']
    df["Date"] = pd.to_datetime(df["Date"], infer_datetime_format=True)   
    result = pd.merge_asof(df,df_env,on="Date")
    result.to_csv(f"../Join/telemetriedaten/positions_weather/{file}")
```
